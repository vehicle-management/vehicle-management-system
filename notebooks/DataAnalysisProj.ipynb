```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

EDA of the dataset car rental sample


```python
car_rental_sample = pd.read_csv('clean_car_rental_sample.csv')

# Analyze rental prices and locations
price_distribution = car_rental_sample['price'].dropna()
top_rental_cities = car_rental_sample['city'].value_counts().head(10)

# ental price distribution
plt.figure(figsize=(8, 6))
price_distribution.plot(kind='hist', bins=20, color='blue', alpha=0.7)
plt.title('Rental Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# top rental cities
plt.figure(figsize=(8, 6))
top_rental_cities.plot(kind='bar', color='orange')
plt.title('Top 10 Cities for Rentals')
plt.xlabel('City')
plt.ylabel('Number of Rentals')
plt.xticks(rotation=45)
plt.show()
```

EDA for the Vehicles Dataset

```python
vehicles = pd.read_csv('clean_vehicles_data.csv')
```

```python
vehicles.info()
```

```python
manufacturer_counts = vehicles['manufacturer'].value_counts().head(10)

# bar chart
manufacturer_counts.plot(kind='bar', color='orange', figsize=(8, 6))
plt.title('Top 10 Vehicle Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot
vehicles.plot.scatter(x='odometer', y='price', alpha=0.5, color='green', figsize=(8, 6))
plt.title('Price vs. Odometer')
plt.xlabel('Odometer (miles)')
plt.ylabel('Price')
plt.grid(alpha=0.3)
plt.show()
```

```python
sns.boxplot(x='condition', y='price', data=vehicles, palette='Set2')
plt.title('Price by Vehicle Condition')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
top_manufacturers = vehicles['manufacturer'].value_counts().head(5).index
top_data = vehicles[vehicles['manufacturer'].isin(top_manufacturers)]

type_counts = top_data.groupby(['manufacturer', 'type']).size().unstack().fillna(0)

# grouped bar chart
type_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title('Vehicle Types by Top Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

```python
yearly_price = vehicles.groupby('year')['price'].mean()

#line chart
yearly_price.plot(kind='line', figsize=(10, 6), color='blue')
plt.title('Average Vehicle Price Over Years')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(alpha=0.3)
plt.show()
```

```python
paint_color_counts = vehicles['paint_color'].value_counts()

# Horizontal bar chart
paint_color_counts.plot(kind='barh', color='purple', figsize=(8, 6))
plt.title('Popular Paint Colors')
plt.xlabel('Count')
plt.ylabel('Paint Color')
plt.show()
```

Car Sales Data

```python
car_sales = pd.read_csv("clean_car_sales_data.csv")
```

```python
sales_data = car_sales
sales_data['Salesperson'].nunique()
```

```python
# Converting 'Date' to datetime and extract month-year
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data['Month-Year'] = sales_data['Date'].dt.to_period('M')

# Group by Month-Year and calculate total sales
monthly_sales = sales_data.groupby('Month-Year')['Sale Price'].sum()

# monthly sales
monthly_sales.plot(kind='line', figsize=(10, 6), color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales ($)')
plt.grid(alpha=0.3)
plt.show()
```

```python
car_make_counts = sales_data['Car Make'].value_counts()

# pie chart
car_make_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 6), startangle=140, colormap='Set3')
plt.title('Car Make Popularity')
plt.ylabel('')  
plt.show()
```

```python
car_model_counts = sales_data['Car Model'].value_counts().head(10)

# bar chart
car_model_counts.plot(kind='barh', color='green', figsize=(8, 6))
plt.title('Top Car Models by Sales Volume')
plt.xlabel('Number of Sales')
plt.ylabel('Car Model')
plt.show()
```

```python
# Scatter plot
sales_data.plot.scatter(x='Sale Price', y='Commission Earned', alpha=0.6, color='purple', figsize=(8, 6))
plt.title('Sale Price vs. Commission Earned')
plt.xlabel('Sale Price ($)')
plt.ylabel('Commission Earned ($)')
plt.grid(alpha=0.3)
plt.show()
```

```python
# Group by car year and count
car_year_counts = sales_data['Car Year'].value_counts().sort_index()

# Plot bar chart
car_year_counts.plot(kind='bar', color='teal', figsize=(8, 6))
plt.title('Number of Cars Sold by Year')
plt.xlabel('Car Year')
plt.ylabel('Number of Sales')
plt.show()
```
